```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the grid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Define the function Z = f(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
# Create the figure
fig = plt.figure(figsize=(7, 3))

# First subplot: basic surface plot
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# Second subplot: surface plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```
