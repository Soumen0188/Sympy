```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5
```

```python
def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
```

```python
phi_m = np.linspace(0, 2*np.pi, 200)
phi_p = np.linspace(0, 2*np.pi, 200)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

```python
fig = plt.figure(figsize=(4, 2))
fig, ax = plt.subplots()
p = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap=plt.cm.plasma, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```

```python
fig = plt.figure(figsize=(6, 4))
fig, ax = plt.subplots()
im = ax.imshow(Z, cmap=plt.cm.magma, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
im.set_interpolation('bilinear')
cb = fig.colorbar(im, ax=ax)
```

```python
fig = plt.figure(figsize=(5, 3))
fig, ax = plt.subplots()
cnt = ax.contour(Z, cmap=plt.cm.inferno, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```

```python
fig = plt.figure(figsize=(10,5))

# `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.viridis, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```
